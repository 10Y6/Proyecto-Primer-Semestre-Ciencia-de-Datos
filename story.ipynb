## 4. ¿Que tan caóticos son los precios?. ¿Es normal las tendencias a subir como ya vimos anteriormente, o estan caracterizadas por picos y valles de precios?

Este análisis es interesante porque hasta ahora hemos visto cuánto han subido los precios , pero no cómo lo han hecho.
Para esto se suele utilizar un concepto estadistico básico, el **Coeficiente de Variación (CV)**, el cual nos permite entender que tan inestable es el mercado.

$$
\text{CV} = \left(\frac{\text{Desviacón Estandar}}{\text{Media Arimética}} \right) \times 100
$$

Matemáticamente hablando nos permite diferenciar entre un patron tendencial, es decir subida o caida constante, o un caos en los precios. 

Para visualizarlo mejor, he clasificado los productos en tres niveles:

* 🟢 Baja Dispersión (CV < 10%): Precios predecibles.
* 🟡 Dispersión Moderada (CV 10-20%): Precios oscilantes.
* 🔴 Alta Dispersión (CV > 20%): Precios caóticos."*



```python
vs.dispersion_analysis()
```

### Análisis de Resultados

#### **Observaciones Generales**
*   La primera impresión que nos podríamos llevar es que ningún producto alcanza niveles de dispersión altos, lo cual nos da algo más de contexto si lo analizamos junto al gráfico #3. Sí, efectivamente la tendencia de los precios es aumentar como veíamos anteriormente, pero aumentan de forma no tan caótica, algo que puede ser causado por la inflación natural.

#### **Observaciones Particulares**

*   **Malta:** El producto más estable de los analizados, presentando un coeficiente de variación de **6.83%**. ¿Qué significa? Las maltas más consumidas en la isla son mayormente de producción nacional; las materias primas nacionales por lo general son más baratas y no dependen tanto de la tasa cambiaria del dólar, lo que reduce costos de producción y, por consecuencia, el precio final al consumidor.

*   **Huevos:** El segundo más estable, aunque cerca del **10%**. Los huevos también son productos de producción nacional, aunque las materias primas para producir los huevos (es decir, piensos específicos para gallinas ponedoras, principalmente hechos de soya y maíz) son importados en su mayoría. Lo que justificaría un coeficiente de variación más alto que la malta, causado por posibles retrasos en la obtención de esta materia prima, lo que desembocaría en el retraso o encarecimiento de producción, finalmente ramificándose al precio a los consumidores.

*   **Aceite:** Lo primero a destacar es que su coeficiente de variación del **10.97%** es más estable de lo que se esperaría de un producto de importación que se vuelve casi dólar-dependiente. Esto podría significar varias cosas, y es que si lo relacionamos con la tasa de cambio, se podría decir que el dólar también posee un coeficiente de variación moderado, regido por el incremento sí, pero sin dar saltos bruscos de forma general.

*   **Cerveza:** Con un coeficiente de variación del **15.36%**, la cerveza se ubica en una zona de dispersión media/alta. Su variación podría deberse a que, a diferencia del arroz o el aceite que son productos indispensables, la cerveza es más un producto de recreación, prescindible para la subsistencia (aunque a todos nos guste la cerveza). Como vimos anteriormente, alcanza picos en fechas festivas como mayo y octubre, y se estabiliza en meses sin muchas festividades.

*   **Arroz:** Aquí ya llegamos a un producto cuanto menos interesante. Cuba importa mayormente el arroz que consume; esto nos podría dar una pista de por qué el arroz se lleva la corona si de coeficiente de variación hablamos (alcanzando un **17.43%**). Al ser un ritual el consumo diario del arroz en los hogares cubanos, es un producto con mucha demanda. Agregando a la fórmula los retrasos en importaciones por disímiles factores, además de que las principales zonas de producción nacional ubicadas en el oriente del país (específicamente en Granma) y dado que la época de cosecha del arroz va desde noviembre hasta febrero principalmente, justo después de la temporada ciclónica, y siendo Granma un territorio muy afectado por huracanes y ciclones, las conclusiones son desoladoras.
