## 2. La "Sombra" del Dolar: ¿Existe Correlación Entre la Tasa de Cambio Informal y el Precio de Distintos Poductos?

A menudo se asume que la inflación en Cuba es un fenómeno puramente monetario: *"Si el dólar sube, todo sube"*. Sin embargo, al analizar los datos recolectados *in situ* y *online* entre Octubre y Noviembre de 2025, la realidad muestra un comportamiento mucho más complejo y especulativo.

En esta sección, correlacionamos la **Tasa de Cambio del Mercado Informal** con el precio promedio diario de dos categorías clave: Víveres y Cárnicos.

> **Objetivo:** Determinar si los precios responden matemáticamente a la tasa de cambio o si existen otros factores que influyen mas en el precio.

### Fuentes y Datos
Los datos han sido procesados a partir de la recopilación de precios de productos in situ en distintas mipymes y publicaciones en grupos de compraventa de facebook, en un periodo que comprende desde finales de octubre a finales de noviembre

###  Empecemos por lo básico: los víveres
Primero analisaremos productos básicos como el aceite, el arroz, el frijol y otros víveres porque son productos que si o si la gente compra en el día a día. Al mirar la gráfica detenidamente, se ven dos momentos que llaman mucho la atención:

```python
vs.corretalion_dollar_product('viveres')
```

*   **La Resistencia a bajar del 30 de octubre al 2 de noviembre:**
    Mirando al inicio, a la izquierda. El dólar cae en picada, pero los víveres se quedan casi con el mismo valor. Esto confirma que aunque el dólar caiga, los vendedores aguantan el precio alto todo lo que pueden antes de bajarlo.

*   **El efecto "Espejo" del 8 al 20 de noviembre:**
    Fíjense en el tramo del medio. Las dos líneas son muy parecidas. El dólar sube y el precio de los víveres sube. Ahí es donde se cumple perfectamente la teoría de que "el dólar manda".      

> **Conclusión:** La gráfica muestra que sí hay conexión, pero que el mercado es muy ineficiente: tiene picos raros que no se explican con dinero y una resistencia enorme a bajar los precios.
